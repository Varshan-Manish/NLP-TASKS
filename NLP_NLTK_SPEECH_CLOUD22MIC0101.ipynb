```python
import nltk
from nltk.corpus import inaugural
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import string
```

```python
nltk.download('inaugural')
```

```python
files = inaugural.fileids()
```

```python
def generate_speech_cloud():
    all_text = ' '.join(inaugural.raw(doc) for doc in inaugural.fileids())

    tokens = word_tokenize(all_text.lower())

    common_words = set(stopwords.words('english'))
    filtered_tokens = [
        token for token in tokens
        if token.isalnum() and token not in common_words and token not in string.punctuation
    ]

    frequency = Counter(filtered_tokens)

    # Use the default font (avoid specifying font_path)
    cloud = WordCloud(
        width=1600,
        height=800,
        background_color='white',
        max_words=100,
        contour_width=3,
        contour_color='steelblue'
    ).generate_from_frequencies(frequency)

    plt.figure(figsize=(20, 10))
    plt.imshow(cloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('USA President Speeches', pad=20)
    plt.show()

    return frequency
```
